```python

def geojson_to_shapely_polygons_original(geojson):
    """
    Convert GeoJSON to a list of tuples containing feature properties and Shapely Polygons.

    Parameters:
    geojson (dict): A dictionary representing the GeoJSON data.

    Returns:
    list: A list of tuples where each tuple is (feature_properties, Shapely Polygon).
    """
    # Create a GeoDataFrame from the GeoJSON
    gdf = gpd.GeoDataFrame.from_features(geojson['features'])
    
    # Extract properties and polygons
    result = []
    for _, row in gdf.iterrows():
        properties = row.to_dict()
        polygon = row['geometry']
        # Remove 'geometry' key from properties
        properties.pop('geometry', None)
        result.append((properties, polygon))
    
    return result


def geojson_to_shapely_polygons_bad(geojson):
    """
    Convert GeoJSON to a list of tuples containing feature properties and Shapely Polygons or MultiPolygons.

    Parameters:
    geojson (dict): A dictionary representing the GeoJSON data.

    Returns:
    list: A list of tuples where each tuple is (feature_properties, Shapely Polygon or MultiPolygon).
    """
    # Create a GeoDataFrame from the GeoJSON
    gdf = gpd.GeoDataFrame.from_features(geojson['features'])
    
    # Extract properties and polygons/multipolygons
    result = []
    for _, row in gdf.iterrows():
        properties = row.to_dict()
        geometry = row['geometry']
        # Remove 'geometry' key from properties
        properties.pop('geometry', None)
        # Check if geometry is Polygon or MultiPolygon
        if isinstance(geometry, (Polygon, MultiPolygon)):
            result.append((properties, geometry))
    
    return result

def geojson_to_shapely_polygons(geojson):
    """
    Convert GeoJSON to a list of tuples containing feature properties and Shapely Polygons.

    Parameters:
    geojson (dict): A dictionary representing the GeoJSON data.

    Returns:
    list: A list of tuples where each tuple is (feature_properties, Shapely Polygon).
    """
    result = []
    
    # Iterate over features in GeoJSON
    for feature in geojson['features']:
        properties = feature['properties']
        geometry = shape(feature['geometry'])
        
        # Check if geometry is Polygon or MultiPolygon
        if isinstance(geometry, Polygon):
            result.append((properties, geometry))
        """
        elif isinstance(geometry, MultiPolygon):
            # this is just wrong
            for polygon in geometry.geoms:
                result.append((properties, polygon))
        """
        
    return result

# Convert GeoJSON to list of tuples
# polygons = geojson_to_shapely_polygons(geojson_example)

# Print the result
# for properties, polygon in polygons:
    # print(f"Properties: {properties}, Polygon: {polygon}")
```

```python
# gdf = gpd.read_file(pathToCountryGeojson)
# gdf.to_file(pathToCountryShapefile)

def geojsonpath2shapefile(geojsonPath, shapefilePath):
    gdf = gpd.read_file(geojsonPath)
    gdf.to_file(shapefilePath)

def geojsonpath2shapes(geojsonPath, shapefilePath):
    geojsonpath2shapefile(geojsonPath, shapefilePath)
    shapes: [fiona.model.Feature] = fiona.open(shapefilePath)
    return shapes

def twogeojsonpaths2shapefile(geojsonPath1, geojsonPath2, shapefilePath):
    gdf1 = gpd.read_file(geojsonPath1)
    gdf2 = gpd.read_file(geojsonPath2)
    combinedgdf = gpd.GeoDataFrame(pd.concat([gdf1, gdf2]))
    combinedgdf.to_file(shapefilePath)

def getAttributeFromShape(shape, attribute):
    result = None
    if shape is None:
        # might want print a warning message here too
        return None
    props = shape["properties"]

    # this method can def be optimized
    
    attr = attribute.upper()
    if (attr in props and props[attr] is not None):
        result = props[attr]
        return result
    attr = attribute.lower()
    if (attr in props and props[attr] is not None):
        result = props[attr]
        return result

    attr = attribute.upper() + "_1"
    if (attr in props and props[attr] is not None):
        result = props[attr]
        return result
    attr = attribute.lower() + "_1"
    if (attr in props and props[attr] is not None):
        result = props[attr]
        
    return result

def getMultiNamedAttributeFromShape(shape, attributes):
    result = None
    for attr in attributes:
        value = getAttributeFromShape(shape, attr)
        if value is not None:
            result = value
            return result

    return result

def getPolygonFromShape(shpe):
    return shape(shpe["geometry"])
    
def getWikidataidFromShape(shpe):
    return getMultiNamedAttributeFromShape(shpe, ["WIKIDATAID", "wikidata_1"])


# polygons: [fiona.model.Feature] = fiona.open(pathToCountryShapefile)

countryShapes = geojsonpath2shapes(pathToCountryGeojson, pathToCountryShapefile)
stateShapes = geojsonpath2shapes(pathToStateGeojson, pathToStateShapefile)

twogeojsonpaths2shapefile(pathToCountryGeojson, pathToStateGeojson, pathToCombinedShapefile)
shapes: [fiona.model.Feature] = fiona.open(pathToCombinedShapefile)

# polygons = countryShapes + stateShapes
polygons = shapes
# polygonsX = shapes

print("Printing polygons...")
# print(polygons)
# for feat in polygons:
    # print(feat)
    # dir(feat)
    # print(feat.keys())
    # print(feat["properties"]["WIKIDATAID"])
    # print(getMultiNamedAttributeFromShape(feat, ["WIKIDATAID", "wikidata_1"]))
    #- print(getWikidataidFromShape(feat))
# dir(polygons)
print("Done printing polygons...")
```

```python
def hex_to_rgba(hex_str):
    """
    ChatGPT made.
    """
    # Remove the '#' character if present
    hex_str = hex_str.lstrip('#')
    
    # Determine the length of the hex string to handle RGB or RGBA
    length = len(hex_str)
    
    if length == 6:  # RGB format
        r, g, b = int(hex_str[0:2], 16), int(hex_str[2:4], 16), int(hex_str[4:6], 16)
        a = 255  # Default alpha value for RGB
    elif length == 8:  # RGBA format
        r, g, b, a = int(hex_str[0:2], 16), int(hex_str[2:4], 16), int(hex_str[4:6], 16), int(hex_str[6:8], 16)
    else:
        raise ValueError("Invalid hex color string length. It should be 6 (RGB) or 8 (RGBA) characters long.")
    
    return (r, g, b, a)

def rgba_to_hex(color):
    """
    ChatGPT made.
    """
    # Ensure the input is a tuple of length 3 (RGB) or 4 (RGBA)
    if len(color) not in [3, 4]:
        raise ValueError("Input must be an RGB tuple of length 3 or an RGBA tuple of length 4.")
    
    # If the input is an RGB tuple, append the alpha value of 255
    if len(color) == 3:
        color = (*color, 255)
    
    # Convert each component to a 2-digit hex string
    return '#{:02X}{:02X}{:02X}{:02X}'.format(color[0], color[1], color[2], color[3])

def hidePlotWhitespace(plt):
    size = (1, 1)
    fig = plt.figure()
    # fig.set_size_inches(size)
    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.add_axes(ax)
    return fig

def saveBitmap(bitmap, path):
    rows = len(bitmap)
    columns = len(bitmap[0])
    print(str(rows) + "x" + str(columns))

    fig = hidePlotWhitespace(plt)
    size = (1, columns / rows)
    fig.set_size_inches(size)

    plt.imshow(bitmap, interpolation='nearest',  cmap="gray")
    plt.savefig(path, bbox_inches='tight', pad_inches = 0, dpi = rows * 2)
    # plt.show()

def replaceColorRGB(inputPath, outputPath, originalColor, replacementColor):
    # orig_color = (255, 255, 255)
    # replacement_color = (128, 128, 0)
    img = Image.open(inputPath).convert('RGB')
    data = np.array(img)
    data[(data == originalColor).all(axis=-1)] = replacementColor
    img2 = Image.fromarray(data, mode='RGB')
    # img2.show()
    img2.save(outputPath)

def replaceColor(inputPath, outputPath, originalColorHex, replacementColorHex):
    # orig_color = (255, 255, 255)
    # replacement_color = (128, 128, 0)
    originalColor = hex_to_rgba(originalColorHex)
    replacementColor = hex_to_rgba(replacementColorHex)
    img = Image.open(inputPath).convert('RGBA')
    data = np.array(img)
    data[(data == originalColor).all(axis=-1)] = replacementColor
    img2 = Image.fromarray(data, mode='RGBA')
    # img2.show()
    img2.save(outputPath)

white = rgba_to_hex((255, 255, 255))
black = rgba_to_hex((0, 0, 0))

lightBlack = rgba_to_hex((27, 27, 27))
gray = rgba_to_hex((136, 136, 136))
lightGray = rgba_to_hex((228, 228, 228))

bitmatOutputPath = 'output/baseboards/pixel-countries.png'
finalOutputPath = 'output/baseboards/pixel-countries-2.png'
saveBitmap(bitmap, bitmatOutputPath)
replaceColor(bitmatOutputPath, finalOutputPath, white, gray)
replaceColor(finalOutputPath, finalOutputPath, black, lightBlack)
```

```python
def simpleReplaceBitmapColor(inputPath, outputPath, waterColor, landColor):
    replaceColor(inputPath, outputPath, white, landColor)
    replaceColor(outputPath, outputPath, black, waterColor)

# def formVariant(inputPath, outputPath, baseboard, waterColor, landColor):

def change_opacity(hex_color, opacity):
    """
    Modify the opacity of a hex color string.

    Parameters:
    hex_color (str): The original hex color string (e.g., '#RRGGBB' or '#RRGGBBAA').
    opacity (float): The new opacity value between 0.0 (completely transparent) and 1.0 (completely opaque).

    Returns:
    str: The modified hex color string with the new opacity.
    """
    hex_color = hex_color.lstrip('#')
    if len(hex_color) not in [6, 8]:
        raise ValueError("Input must be a hex color string of length 6 (RGB) or 8 (RGBA).")
    rgb = hex_color[:6]
    a = int(opacity * 255)
    new_hex_color = f"#{rgb}{a:02X}"
    return new_hex_color


TRANSPARENT_HEX = "#00000000"
HIGHLIGHT_OPACITY = 0.5
DELIM = "_"

def drawAllTerritories(tileOwners, allWikidataids):
    """
    Draw all territories. Could potentially later be used to make variants of countries in different colors.
    """
    # pathToVariantOutputFolder
    # f it imma just go fot it
    # iterate over all wikidataids
    for wikidataid in allWikidataids:
        # optional create an empty text file with the long name of the country here inside the folder
        countryFolder = pathToVariantOutputFolder + wikidataid + "/"
        os.makedirs(countryFolder, exist_ok=True)
        bitmapOutputPath = countryFolder + "BITMAP.png"
        drawCountryFromTileOwners(tileOwners, wikidataid, bitmapOutputPath)

        # for baseboard in Baseboard.values():
        outputPath = countryFolder + "TRANSPARENT" + DELIM + "GRAY" + ".png"
        simpleReplaceBitmapColor(bitmapOutputPath, outputPath, TRANSPARENT_HEX, colorname2hex["gray"])
        outputPath = countryFolder + "TRANSPARENT" + DELIM + "YELLOW" + ".png"
        simpleReplaceBitmapColor(bitmapOutputPath, outputPath, TRANSPARENT_HEX, colorname2hex["yellow"])
        outputPath = countryFolder + "TRANSPARENT" + DELIM + "TRANSLUCENT_WHITE" + ".png"
        simpleReplaceBitmapColor(bitmapOutputPath, outputPath, TRANSPARENT_HEX, change_opacity(colorname2hex["white"], HIGHLIGHT_OPACITY))
        outputPath = countryFolder + "TRANSPARENT" + DELIM + "TRANSLUCENT_LIGHT_GRAY" + ".png"
        simpleReplaceBitmapColor(bitmapOutputPath, outputPath, TRANSPARENT_HEX, change_opacity(colorname2hex["lightGray"], HIGHLIGHT_OPACITY))
        outputPath = countryFolder + "TRANSPARENT" + DELIM + "TRANSLUCENT_GRAY" + ".png"
        simpleReplaceBitmapColor(bitmapOutputPath, outputPath, TRANSPARENT_HEX, change_opacity(colorname2hex["gray"], HIGHLIGHT_OPACITY))
        outputPath = countryFolder + "TRANSPARENT" + DELIM + "TRANSLUCENT_BLACK" + ".png"
        simpleReplaceBitmapColor(bitmapOutputPath, outputPath, TRANSPARENT_HEX, change_opacity(colorname2hex["black"], HIGHLIGHT_OPACITY))
        outputPath = countryFolder + "TRANSPARENT" + DELIM + "TRANSLUCENT_YELLOW" + ".png"
        simpleReplaceBitmapColor(bitmapOutputPath, outputPath, TRANSPARENT_HEX, change_opacity(colorname2hex["yellow"], HIGHLIGHT_OPACITY))
        

allWikidataids = getAllWikidataids()
drawAllTerritories(tileOwners, allWikidataids)
```

```python
from PIL import Image

assetFolder = "../app/assets/"
assetHighlightFolder = assetFolder + "/highlights/"

'''
def compress_pngs_in_folder(folder_path, max_depth=2, quality=10, optimize=True): # quality=85
    """
    Compress all PNG files in the given folder and its subfolders up to a specified depth.

    Parameters:
    folder_path (str): The path to the folder containing PNG files.
    max_depth (int): The maximum depth of subfolders to traverse.
    quality (int): The quality setting for the output images.
    optimize (bool): Whether to optimize the images.
    """
    folder_path = os.path.abspath(folder_path)  # Ensure folder path is absolute

    # Walk through all directories and subdirectories
    for root, dirs, files in os.walk(folder_path):
        # Calculate the current depth by counting the number of directories from the root
        current_depth = root[len(folder_path):].count(os.sep)
        if current_depth > max_depth:
            raise ValueError(f"Folder structure is too deeply nested. Maximum allowed depth is {max_depth}.")

        for filename in files:
            # Check if the file is a PNG file
            if filename.lower().endswith('.png'):
                file_path = os.path.join(root, filename)
                print(f"Compressing {file_path}")
                
                # Open the PNG file
                img = Image.open(file_path)
                
                # Save the image with the specified compression settings
                img.save(file_path, format='PNG', optimize=optimize, quality=quality)
                print(f"Compressed {file_path}")
'''
import subprocess

def compress_png_with_optipng(file_path, optimization_level=2):
    """
    Compress PNG file using optipng.

    Parameters:
    file_path (str): Path to the PNG file to be compressed.
    optimization_level (int): Optimization level (0-7). Higher means better compression.
    """
    command = [
        'optipng', '-o', str(optimization_level), file_path
    ]

    try:
        subprocess.run(command, check=True)
        print(f"Compressed {file_path}")
    except subprocess.CalledProcessError as e:
        print(f"Error compressing {file_path}: {e}")

def compress_pngs_in_folder(folder_path, optimization_level=2):
    """
    Compress all PNG files in the given folder and its subfolders using optipng.

    Parameters:
    folder_path (str): The path to the folder containing PNG files.
    optimization_level (int): Optimization level (0-7).
    """
    for root, dirs, files in os.walk(folder_path):
        for filename in files:
            if filename.lower().endswith('.png'):
                file_path = os.path.join(root, filename)
                compress_png_with_optipng(file_path, optimization_level)


# compress_pngs_in_folder(countryFolder)
# compress_pngs_in_folder(assetHighlightFolder)
```
